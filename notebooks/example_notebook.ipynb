# Example Jupyter Notebook

This notebook demonstrates basic calculations and visualization using Python.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Basic Calculations

```python
# Perform some basic calculations
a = 10
b = 5

print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
```

## Using NumPy for Array Operations

```python
# Create arrays with NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

print(f"Array x: {x[:5]}...")
print(f"Array y: {y[:5]}...")
print(f"Mean of y: {np.mean(y):.4f}")
print(f"Std of y: {np.std(y):.4f}")
```

## Visualization with Matplotlib

```python
# Create a simple plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.plot(x, np.cos(x), 'r--', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions')
plt.legend(['sin(x)', 'cos(x)'])
plt.grid(True)
plt.show()
```

## Using Pandas for Data Analysis

```python
# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'Score': [85, 92, 78, 95, 88]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print(f"\nAverage Score: {df['Score'].mean():.2f}")
```

## Bar Chart Visualization

```python
# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Student Scores')
plt.ylim(0, 100)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
